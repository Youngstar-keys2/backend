```python
columns = ['Коды ТН ВЭД ЕАЭС', 
           'Технические регламенты', 
           'Общее наименование продукции', 
           'Группа продукции']


def split_series(series: pd.Series) -> pd.DataFrame:
  good = True
  for column in columns:
    if ';' in str(series[column]):
      good = False
  if good:
    return pd.DataFrame([series])
  else:
    res = []

    id = series['Номер продукции']
    name = series['Общее наименование продукции']
    il = series['ИЛ']
    zav = series['Заявитель']    
    adr_zav = series['Адрес Заявителя']
    izg = series['Изготовитель']
    country = series['Страна']
    adr_izg = series['Адрес изготовителя']


    

    tnveds = str(series['Коды ТН ВЭД ЕАЭС']).split(';')
    tehs = str(series['Технические регламенты']).split(';')
    groups = str(series['Группа продукции']).split(';')

    tnveds = [t.strip() for t in tnveds]
    tehs = [t.strip() for t in tehs]
    groups = [t.strip() for t in groups]
    
    for tnved in tnveds:
      for teh in tehs:
        for group in groups:
          res.append(pd.Series(data=[id, tnved, teh, group, name, il, zav, adr_zav, izg, country, adr_izg], index=df.columns))
          # print(tnved, teh, group)
    # print(tnveds)
    # print(tehs)
    # print(groups)
    res = pd.DataFrame(data=res,columns=df.columns)
    return res
    
# sertest = pd.Series(data=['1234567890; 0987654321','a; b;c', 'test object', 'd;e;f'], index=columns)
sertest = pd.Series(data=['кок продукция', '1234567890; 0987654321','a; teh','group', 'test object', 'ил', 'заяв', 'адрес заявителя', 'изготовитель','страна' ,'адрес изготовителя'], index=df.columns)
# sertest = pd.Series(data=['1234567890','c', 'test object 2', 'f'], index=columns)
print(sertest)
res = split_series(sertest)
res
```

```python

# %%time

# Drop duplicates
df_for_model = df_for_model.dropna().drop_duplicates()

# Remove quotes and convert to lower
df_for_model['Общее наименование продукции'] = df_for_model['Общее наименование продукции'].parallel_apply(lambda x: re.sub("'", '', x).lower())

# Remove all the special chars
exclude = set(string.punctuation) # set of all special chars
df_for_model['Общее наименование продукции'] = df_for_model['Общее наименование продукции'].parallel_apply(lambda x: ''.join(ch for ch in x if ch not in exclude))

# Remove digits
remove_digits = str.maketrans('','',''.join([str(x) for x in range(0,10)]))
df_for_model['Общее наименование продукции'] = df_for_model['Общее наименование продукции'].parallel_apply(lambda x: x.translate(remove_digits))

# Remove extra spaces
df_for_model['Общее наименование продукции'] = df_for_model['Общее наименование продукции'].parallel_apply(lambda x: x.strip())
df_for_model['Общее наименование продукции'] = df_for_model['Общее наименование продукции'].parallel_apply(lambda x: re.sub(" +", " ", x))

# Стемминг
# from nltk.stem import SnowballStemmer
# ss = SnowballStemmer(language='russian')
# df_for_model['Общее наименование продукции'] = df_for_model['Общее наименование продукции'].parallel_apply(lambda x: ' '.join(list(map(ss.stem, x.split(' ')))))

# #TODO: Лемматизация!!! 
nlp = spacy.load("ru_core_news_lg")
i = 0
def lemmatize(x):
  # if i % 1000:
    # print(i)
  # i += 1
  lemma_list = []
  for token in x:
      lemma_list.append(token.lemma_)
  normalized_tokens = [] 
  for word in lemma_list:
      lexeme = nlp.vocab[word]
      if lexeme.is_stop == False:
          normalized_tokens.append(word)
  return normalized_tokens
  
df_for_model['Общее наименование продукции'] = df_for_model['Общее наименование продукции'].apply(lambda x: ' '.join(lemmatize(nlp(x))))

# обрезаем до 8 слов
df_for_model['Общее наименование продукции'] = df_for_model['Общее наименование продукции'].parallel_apply(lambda x: ' '.join(x.split(' ')[:8]))



# pymorphy_results = list(map(lambda x: morph.parse(x), raw_ru.split(' ')))
# print(' '.join([res[0].normal_form for res in pymorphy_results]))
# И разделим коды ТН ВЭД ЕАЭС на три части
# df_for_model['GRUPPA'] = df['Коды ТН ВЭД ЕАЭС'].parallel_apply(lambda x: x[0:2])
# df_for_model['TOV_POZ'] = df['Коды ТН ВЭД ЕАЭС'].parallel_apply(lambda x: x[2:4])
# df_for_model['SUB_POZ'] = df['Коды ТН ВЭД ЕАЭС'].parallel_apply(lambda x: x[4:])

print(df_for_model.shape)
df_for_model.head()
```

```python
from pprint import pprint

df_temp = pd.DataFrame(df).drop('Номер продукции', axis=1)

def predict_pair(name_nlp, second_nlp):
  doc1 = name_nlp
  doc2 = second_nlp
  res = doc1.similarity(doc2)
  # print(doc1, ' <=> ', doc2, ' = ', res)
  return res

def predict(name: str, df: pd.DataFrame) -> float:
  name_nlp = nlp(name)
  df_names = df['vec'].values
  sims = []
  i = 0
  for name in df_names:
    sims.append((i, predict_pair(name_nlp, name)))
    i += 1
  sims = sorted(sims, key=lambda x: x[1], reverse=True)
  print(sims[:5])
  tnved_dict = {}
  tehreg_dict = {}
  group_dict = {}
  COUNT = 3
  i = 0
  for sim in sims[:COUNT]:
    sample = df_temp.values[sim[0]]
    mult = sim[1]
    score = mult
    tnved_dict[sample[0]] = score if not sample[0] in tnved_dict else tnved_dict[sample[0]]  + score
    tehreg_dict[sample[1]] = score if not sample[0] in tehreg_dict else tehreg_dict[sample[0]] + score
    group_dict[sample[2]] = score if not sample[0] in group_dict else group_dict[sample[0]] + score
    i += 1
  return {
      "tnved": max(tnved_dict, key=tnved_dict.get),
      "tehreg": max(tehreg_dict, key=tehreg_dict.get),
      "group": max(group_dict, key=group_dict.get),
  }, df_temp.values[sims[0][0]], df_temp.values[sims[1][0]], df_temp.values[sims[2][0]], df_temp.values[sims[3][0]] 

  # return 
stem = lambda x: ' '.join(list(map(ss.stem, x.split(' '))))
pred = stem('ноутбук'.lower())
print(pred)
pprint(predict(pred, df))
```

```python
def validate(x, df):
  pred = predict(x['name'], df)[0]
  # print("Prediction:")
  # pprint(pred)
  return x['tnved'] == pred['tnved'] and x['tehreg'] == pred['tehreg'] and x['group'] == pred["group"]

pred = predict('ноутбук', df)[1]
pred_dict = {
    'tnved': pred[0],
    'tehreg': pred[1],
    'group': pred[2],
    'name': pred[3],
}
pprint(pred_dict)
print(validate(pred_dict, df))
pred_dict['name'] = 'смартфон'
pprint(pred_dict)

print(validate(pred_dict, df))
```
